```python
#import time
import pandas as pd

list = []
f = open('MachineData.log')
line = f.readline()

while line:  

    try:

        list.append(line)               

    except ValueError:

        print('Error in line :' + line )

    line = f.readline()
```

```python
list
```

```python
df = pd.DataFrame([sub.split(" : ") for sub in list])
df.columns = ['DateAndTime', 'Description']
df.head()
```

```python
DateTimeSplit = df["DateAndTime"].str.split(" ", n = 1, expand = True)
```

```python
df["DT"]= DateTimeSplit[0]
df["Time"]= DateTimeSplit[0]
df.head()
```

```python
df['Details']=df.Description.str.extract('\<(.*)\>')
```

```python
new = df["Description"].str.split(">", n = 1, expand = True)
df["RobotDetails"]= new[1]
df.head()
```

```python
df['Date'] = pd.to_datetime(df['DateAndTime'], dayfirst=True).dt.strftime('%m.%d.%Y')
df['Day'] = pd.to_datetime(df['DateAndTime'], dayfirst=True).dt.strftime('%d')
df['Month'] = pd.to_datetime(df['DateAndTime'], dayfirst=True).dt.strftime('%m')
df['Year'] = pd.to_datetime(df['DateAndTime'], dayfirst=True).dt.strftime('%Y')
df['Time'] = pd.to_datetime(df['DateAndTime'], dayfirst=True).dt.time
df['CW'] = pd.to_datetime(df['DateAndTime'], dayfirst=True).dt.strftime('%W')
df['DT'] = pd.to_datetime(df['DateAndTime'])
df.head()
```

```python
import re
deliverpo = []
fromlist = []
tolist = []
itemlist = []
polist = []
reg = re.compile('Deliver PO')
#Regular expressions are compiled into pattern objects, which have methods for various operations such as searching for 
#pattern matches or performing string substitutions.

# remember that column RobotDetails looks e.g. like this: Deliver PO: 1212201 (Item I2345672) by area:10 to area: 1
for index, row in df.iterrows():
    
    SearchMe = re.search(reg, row['RobotDetails'])
# each row will be searched for the string Deliver PO
    if SearchMe is not None:
        # When found Deliver PO in one row, then take the number out of the 3rd split in between the : separated column RobotDetails
        # that number is eihter a 1 followed by 0 or 1 or 2, or it is just one number between 0-9
        # for the first line in our log file example this would be 9 as
        by = re.findall('1[012]|[0-9]', re.split('[:]', row['RobotDetails'])[2]).pop()
        # When found Deliver PO in one row, then take the number out of the 4th split in between the : separated column RobotDetails
        # that number is eihter a 1 followed by 0 or 1 or 2, or it is just one number between 0-9
        to = re.findall('1[012]|[0-9]', re.split('[:]', row['RobotDetails'])[3]).pop()
        # to get the order number, 
        # \d matches any decimal digit; this is equivalent to the class [0-9]. So within RobotDetails we are just looking for an order number the size of 7 digits
        order = re.findall('\d{7}', row['RobotDetails']).pop(0)
        # \d matches any decimal digit; this is equivalent to the class [0-9]. So within RobotDetails we are just looking for an item number the size of 6 digits and pop it out second (1) as series
        item = re.findall('\d{6}', row['RobotDetails']).pop(1)
        
        fromlist.append(by)
        tolist.append(to)
        print(order, item)
        itemlist.append('I'+item)
        polist.append(order)
        
    else:
        fromlist.append(None)
        tolist.append(None)
        itemlist.append(None)
        polist.append(None)
    deliverpo.append(SearchMe)    


df['FromPosition'] = fromlist
df['ToPosition'] = tolist
df['ProductionOrder'] = polist
df['Item'] = itemlist
```

```python
from datetime import datetime, timedelta 
deltaT = []
#start time
t1 = None
#end time
t2 = None

# must be instantiated with 1 as a  starting point
toggle = 1

FromBucket = None
ToBucket = None


for index, row in df.iterrows():
    by = row['FromPosition'] 
    to = row['ToPosition']
    # not equal 0 is written like that: != '0'
    if by is not None and to is not None and by != '0' and to != '0':
        
        if toggle: 
            #Toggle: Fill in values for the first time
            t1 = row['DT']
            FromBucket = by
            ToBucket = to
            toggle = 0
            deltaT.append(None)
            
        else:
            # Check if first order has been closed (forth and back)
            # That's the interesting part: we only want to calculate the put trough time if one order has been closed successfully
            # So while in the loop we are checking for matching "to" and "from" orders and only calculate the time difference for those
            if FromBucket == to and ToBucket == by:
                # if yes, then take End time and calculate time difference
                t2 = row['DT']
                timedelta = t2-t1
                deltaT.append(timedelta)
                
            else: 
                # if not, fill in error with current value
                t1 = row['DT']
                FromBucket = by
                ToBucket = to
                deltaT.append(None)
        print(timedelta)
    else: 
        deltaT.append(None)
        
df['deltaT'] = deltaT
```
